### Q1 : if man is to king then what will be woman is ______________?

```python
# vector('man') and vector('king') -> Relationship --> women ---> Prediction
```

```python
result = model.most_similar(positive=['woman','king'],negative=['man'])
```

```python
answer,score = result[0]
```

```python
print('Answer with score :',answer,score)
```
